# Matplotlib Basics

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = np.arange(0,10)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
#plt.xlabel("x-axis")
plt.xlim(0,6)
plt.ylim(0,15)
plt.title("String Title")
plt.savefig("plot.png")
```

# Matplotlib - Figures

```python
a = np.linspace(0,10,11)
a
```

```python
b = a**4
b
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
# large axes
axes_1 = fig.add_axes([0,0,1,1])
axes_1.plot(a,b)
axes_1.set_xlim(0,8)
axes_1.set_ylim(0,8000)
axes_1.set_xlabel("A")
axes_1.set_ylabel("B")
axes_1.set_title("power of 4")

# small axes 
axes_2 = fig.add_axes([0.2,0.2,0.4,0.4])
axes_2.plot(x,y)
axes_2.set_xlim(0,8)
axes_2.set_ylim(0,15)
axes_2.set_xlabel("x")
axes_2.set_ylabel("y")
axes_2.set_title("multiply of 2")



;
```

# Matplotlib - Figure 

```python
fig = plt.figure(dpi=100)
axes = fig.add_axes([0,0,1,0.5])
axes.plot(x,y)
fig.savefig("new_figure.png",bbox_inches="tight");
```

# Matplotlib - Subplots 

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
plt.tight_layout();
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
fig.subplots_adjust(wspace=0.5,hspace=0.5)
fig.suptitle("Figure level",fontsize=16)
fig.savefig("new_subplot.png",bbox_inches = "tight")
```

# Matplotlib - Adding  Legends

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x,label = "x vs x")
axes.plot(x,x**2,label="x vs x**2")
axes.legend()
```

```python
fig = plt.figure(dpi=200)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x,label = "x vs x")
axes.plot(x,x**2,label="x vs x**2")
axes.legend(loc=(1.1,0.5))
```

# Matplotlib - line  Colors and Style

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x,color="red",label="X vs X")
axes.plot(x,x+1,color="#e6c017",label="X vs X+1")
axes.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x,color="red",label="X vs X",lw=5,linestyle="-.")
axes.plot(x,x+1,color="#e6c017",label="X vs X+1",linewidth=2,marker="o",markersize=10,markerfacecolor="red"
         ,markeredgewidth=2,markeredgecolor="purple")#linestyle="--",
axes.legend()
```
